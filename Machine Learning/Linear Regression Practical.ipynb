# what are the assumption in linear regression?


.We do assumption to validate our model.


steps:
    1. we do scatter plot between predicted value and y_test value and if the model is linear,
    then the model is good.
    2. We caculate residual, and plot it in displot , if the distribution is normally distributed, then we can say our model is good.
    3. Scatter plot of y_pred and resiudal must be uniformly distributed

```python
plt.scatter(reg_pred, y_test)
plt.xlabel("Predicted data")
plt.ylabel("Test data")
```

```python
#calcute the residual
residuals = y_test - reg_pred
```

```python
residuals
```

```python
sns.displot(residuals, kind = 'kde')
```

```python
#scatter plot with prediction and residual
#uniform distribution
plt.scatter(reg_pred,residuals)
```

```python
##performance metrics
#Homoscedasticity, or homogeneity of variances, is an assumption of equal or similar variances in different groups being compared. This is an important assumption of parametric statistical tests because they are sensitive to any dissimilarities. Uneven variances in samples result in biased and skewed test results.
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mean_absolute_error(y_test, reg_pred)
```

```python
np.sqrt(mean_absolute_error(y_test, reg_pred))
```

```python
##R-Square and adjusted R-Square
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test, reg_pred)
```

```python
score
```

```python
#Adjusted R-Square
```

```python
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
